```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('jobs_final.csv')
data.head()
```

```python
#Some information about the data.
data.info()
# Here we can see basic statistics like mean and quartiles about the two integer columns in the dataset.
# We can also see that the highest and lowest number of the salaries.
data.describe()
#To categorical data
#data = data.astype({'Location':'category','Job-title':'category','Job-level':'category','Company size':'category'})
print("Done!")
```

```python
#First is how many n unique values are there in each column of the dataframe
print(data.nunique())
#Now for each Column.
print(data.Company.value_counts())
print(data.Location.unique())
print(data['Job-title'].unique())
print(data['Job-level'].unique())
print(data["Expected Salary lpa"].unique())
print(data['Company size'].unique())
print(data['No of vacancies'].unique())
```

# Step : Exploratory Data analysis.

# Finding Relations

```python
# Location vs Expected salary
sns.set(rc={"figure.figsize":(8, 5)})
sns.scatterplot(data = data, x = 'Expected Salary lpa', y = 'Location', hue= 'Expected Salary lpa')
```

```python
# Company Size vs Expected Salary
sns.set(rc={"figure.figsize":(8, 5)})
sns.scatterplot(data = data, x = 'Expected Salary lpa', y = 'Company size', hue = 'Expected Salary lpa')
```

```python
# Job title vs Expected Salary
sns.set(rc={"figure.figsize":(12, 7)})
sns.histplot(data = data, x = 'Expected Salary lpa', y = 'Job-title', hue = 'Expected Salary lpa')
```

```python
# Job-level vs Expected Salaries. 
sns.set(rc={"figure.figsize":(12, 8)})
sns.histplot(data = data, x = 'Expected Salary lpa', y = 'Job-level', hue = 'Expected Salary lpa')
```

```python
#Job level vs No of vacancies
sns.set(rc={"figure.figsize":(12,7)})
sns.displot(data = data, x = 'No of vacancies', y = 'Job-level', hue = 'No of vacancies')
```

```python
#Job title vs No of vacancies
sns.set(rc={"figure.figsize":(12, 8)})
sns.displot(data = data, x = 'No of vacancies', y = 'Job-title', hue = 'No of vacancies')
```

```python
#Company Size vs No of vacancies
sns.set(rc={"figure.figsize":(12, 7)})
sns.displot(data = data, x = 'No of vacancies', y = 'Company size', hue = 'No of vacancies')
```

```python
#Location vs No of vacancies
sns.set(rc={"figure.figsize":(12, 7)})
sns.displot(data = data, x = 'No of vacancies', y = 'Location', hue = 'No of vacancies')
```
